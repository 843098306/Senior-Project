2. Use stationary distribution to find combination of lambda and mue that minimizes waiting time.

Function: **calculatewaittimelist**  
Description: Given the lambda value, iterate through different mu values and record waiting time of each combination. (fixed lambda value)
Input: (lambdav)  
Output: temp  
See Senior_Project_Function.py for detailed information

Example Usage: **calculatewaittimelist**(0.5)

Find best $\mu$ values for situations with certain $\lambda$ values by hard coding.

```python
DataWait = pd.DataFrame(columns= ["mu1","mu2","mu3","mu4","WaitingTimeTotalmu1","WaitingTimeTotalmu2","WaitingTimeTotalmu3","WaitingTimeTotalmu4"])
temp1 = calculatewaittimelist(0.5)
temp2 = calculatewaittimelist(1)
temp3 = calculatewaittimelist(1.5)
temp4 = calculatewaittimelist(2)

for i in range(5,36):
    DataWait = DataWait.append({"mu1":1/i,"mu2":1/i,"mu3":1/i,"mu4":1/i,
                                "WaitingTimeTotalmu1":temp1[i - 5],
                                "WaitingTimeTotalmu2":temp2[i - 5],
                                "WaitingTimeTotalmu3":temp3[i - 5],
                                "WaitingTimeTotalmu4":temp4[i - 5]}, ignore_index=True)
    
fig, ax = plt.subplots()
ax.scatter(x=DataWait.iloc[:,0],y=DataWait.iloc[:,4],c='DarkBlue',label = "lambda = 0.5")
ax.scatter(x=DataWait.iloc[:,1],y=DataWait.iloc[:,5],c='Green',label = "lambda = 1")
ax.scatter(x=DataWait.iloc[:,2],y=DataWait.iloc[:,6],c='Red',label = "lambda = 1.5")
ax.scatter(x=DataWait.iloc[:,3],y=DataWait.iloc[:,7],c='Pink',label = "lambda = 2")
plt.title(r'Expected Total Waiting List with Different $ \lambda$')
plt.xlabel(r'$\mu$')
plt.ylabel('Expected Total Waiting List')
ax.legend()
path = 'Pictures'
p = str(path + "\lambda(0.5+1+1.5+2)WaitingTimeTotal.png")
plt.savefig(p)
```
